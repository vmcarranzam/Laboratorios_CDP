4. ¿Cuál es el municipio situado más al Norte? (Usar el

5. Obtenga la media, mediana, desviación estándar, valor

```python
df.loc[df['Provincia'] == 'Granada'].describe()
```

6. Realice un histograma con la población de los
Municipios para cada una de las provincias.

```python
for provincia in df["Provincia"].unique():
    df_provincia = df[df["Provincia"] == provincia]
    plt.hist(df_provincia["Poblacion"])
    plt.xlabel("Población")
    plt.ylabel("Número de habitantes")
    plt.title("población de los Municipios de {}".format(provincia))
    plt.show()
```
