###Calculando el coeficiente de correlaciÃ³n de cada variable independiente

```python
corr= [df.SalePrice.corr(df.OverallQual),df.SalePrice.corr(df.FirstFlrSF),df.SalePrice.corr(df.TotRmsAbvGrd),df.SalePrice.corr(df.YearBuilt),df.SalePrice.corr(df.LotFrontage)]

print("CorrelaciÃ³n entre SalePrice y OverallQual es {} %".format(corr[0]*100))
print("CorrelaciÃ³n entre SalePrice y FirstFlrSF es {} %".format(corr[1]*100))
print("CorrelaciÃ³n entre SalePrice y TotRmsAbvGrd es {} %".format(corr[2]*100))
print("CorrelaciÃ³n entre SalePrice y YearBuilt es {} %".format(corr[3]*100))
print("CorrelaciÃ³n entre SalePrice y LotFrontage es {} %".format(corr[4]*100))
```
